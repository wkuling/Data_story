```python
# Let's set the rows (index) first...
temp = investments
temp = investments.groupby(['company_category_code', 'funded_year']).sum()
temp = temp['raised_amount_usd']
temp = temp.divide(10**9)
```

```python
temp=temp.unstack()
temp
#temp=np.round(temp,2)
```

```python
temp['Total']=temp.sum(axis=1)
```

```python
temp.drop(temp.columns[:15], axis=1, inplace=True)
```

```python
temp=temp.sort('Total', ascending=False)
```

```python
temp.drop(temp.index[10:], axis=0, inplace=True)
temp.drop('Total', axis=1, inplace=True)
```

```python
temp.fillna(0, inplace=True)
temp=np.round(temp,2)
temp

```

```python
temp.columns
```

```python
temp.columns = range(2004,2015)
```
